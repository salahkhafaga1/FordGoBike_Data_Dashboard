```python
import numpy as np
import matplotlib.pyplot as plt

# وسعنا النطاق الزمني عشان يغطي عمليات التشفيت (Shift) براحتنا
n = np.arange(-10, 15)

# تعريف الدوال الأساسية (Step & Impulse)
def delta(n_val): return np.where(n_val == 0, 1.0, 0.0)
def u(n_val): return np.where(n_val >= 0, 1.0, 0.0)

# تعريف الإشارات الأساسية كدوال (Functions) لتسهيل التعويض
def f1(n_val): return 3*delta(n_val+1) - 2*delta(n_val) + delta(n_val-1) + 0.5*delta(n_val-2)
def f2(n_val): return 10*u(n_val) + 5*u(n_val-5)
def f3(n_val): return u(n_val) - u(n_val-8)
def f4(n_val): return (2.0**n_val) * (u(n_val+3) - u(n_val-3))
def f5(n_val): return ((0.5)**n_val) * (u(n_val+2) - u(n_val-3))

# حساب الإشارات المطلوبة في السؤال
f6 = 10 * f1(n)
f7 = (1/5) * f2(n)
f8 = f3(n) + f4(n)
f9 = f4(n) - f5(n)
f10 = f3(2*n + 5)
f11 = f4(-n + 2)
f12 = 10 * f5(n - 3)

# تجهيز لوحة الرسم لـ 7 رسومات تحت بعض
fig, axes = plt.subplots(7, 1, figsize=(12, 25))

# تجميع الإشارات وعناوينها في قائمة لتسهيل الرسم
signals = [
    (f6, 'f) f6(n) = 10 f1(n)'),
    (f7, 'g) f7(n) = (1/5) f2(n)'),
    (f8, 'h) f8(n) = f3(n) + f4(n)'),
    (f9, 'i) f9(n) = f4(n) - f5(n)'),
    (f10, 'j) f10(n) = f3(2n+5)'),
    (f11, 'k) f11(n) = f4(-n+2)'),
    (f12, 'l) f12(n) = 10 f5(n-3)')
]

# رسم كل إشارة في مكانها
for ax, (sig, title) in zip(axes, signals):
    ax.stem(n, sig)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.set_xticks(np.arange(-10, 16, 1)) # إظهار كل الأرقام على محور x
    ax.set_ylabel('Amplitude')

plt.xlabel('n (Time Index)', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# تحديد النطاق الزمني n 
n = np.arange(-8, 9)

# 1. تعريف الإشارة الأصلية x(n) من الرسم
def x_func(n_val):
    # الدالة بتاخد قيمة مطلقة لـ n في الفترة من -3 لـ 3، وغير كده بصفر
    cond = (n_val >= -3) & (n_val <= 3)
    return np.where(cond, np.abs(n_val), 0)

# 2. تعريف الإشارة الأصلية y(n) من الرسم
def y_func(n_val):
    # الدالة بـ 1 في الموجب (1 لـ 4) وبـ -1 في السالب (-1 لـ -4)
    cond_pos = (n_val >= 1) & (n_val <= 4)
    cond_neg = (n_val >= -4) & (n_val <= -1)
    return np.select([cond_pos, cond_neg], [1, -1], default=0)

# 3. حساب الإشارات المطلوبة في السؤال
sig1 = x_func(2 * n)
sig2 = x_func(3 * n - 1)
sig3 = y_func(-2 * n + 2)
sig4 = x_func(-n) * y_func(-n)

# 4. تجهيز لوحة الرسم
fig, axes = plt.subplots(4, 1, figsize=(10, 16))

# تجميع الإشارات وعناوينها
signals = [
    (sig1, '1) x(2n)'),
    (sig2, '2) x(3n - 1)'),
    (sig3, '3) y(-2n + 2)'),
    (sig4, '4) x(-n) * y(-n)')
]

# رسم كل إشارة
for ax, (sig, title) in zip(axes, signals):
    ax.stem(n, sig)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.set_xticks(np.arange(-8, 9, 1)) # إظهار الأرقام بوضوح على محور x
    ax.set_yticks(np.arange(-4, 5, 1)) # توحيد محور y عشان الرسمة تبقى متناسقة
    ax.set_ylabel('Amplitude')

plt.xlabel('n (Time Index)', fontsize=14)
plt.tight_layout()
plt.show()
```
